```python
movies_df = df[[ 'title', 'director', 'cast', 'description', 'country', 'rating','listed_in']]
movies_df.head()
```

```python
movies_df.isnull().sum()
```

```python
movies_df['director'] = movies_df['director'].fillna('')
movies_df['cast'] = movies_df['cast'].fillna('')
movies_df['country'] = movies_df['country'].fillna('')
movies_df['rating'] = movies_df['rating'].fillna('')
```

```python
movies_df.isnull().sum()
```

```python
movies_df['cast'] = movies_df['cast'].apply(lambda x:x.split(", ")[0:10])
```

```python
movies_df.cast.head()
```

```python
movies_df['description'] = movies_df['description'].apply(lambda x:x.split())
movies_df['rating'] = movies_df['rating'].apply(lambda x:x.split())
movies_df['country'] = movies_df['country'].apply(lambda x:x.split(","))
movies_df['director'] = movies_df['director'].apply(lambda x:x.split(","))
movies_df['listed_in'] = movies_df['listed_in'].apply(lambda x:x.split(","))
```

```python
movies_df['features'] = movies_df['cast']+movies_df['description']+movies_df['rating']+movies_df['country']+movies_df['director']+movies_df['listed_in']
```

```python
feature_df = movies_df[['title', 'features']]
```

```python
feature_df.head()
```

```python
feature_df['features']= feature_df['features'].apply(lambda x : list(filter(None, x)))
```

```python
feature_df['features']=feature_df['features'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
feature_df['features']=feature_df['features'].apply(lambda x : " ".join(x))
```

```python
feature_df['features']=feature_df['features'].apply(lambda x : x.lower())
```

```python
feature_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(feature_df['features']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
movie = 'Jaane Tu... Ya Jaane Na'
```

```python
movie_index = feature_df[feature_df['title'] == movie].index[0]
```

```python
similar_movies = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])
similar_movies[0:6]
```

```python
for i in similar_movies[1:6]:
        print(feature_df.iloc[i[0]].title)
```
